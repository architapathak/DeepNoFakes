# DNF 300

```python
def load_spacy():
    sentencizer = English()
    sentencizer.add_pipe(sentencizer.create_pipe('sentencizer'))
    nlp = spacy.load("en_core_web_md")
    return sentencizer, nlp
sentencizer, nlp = load_spacy()
```

```python
dnf300_1 = pd.read_json('../evaluation_set/deepnofakes/TRAIN/Set1_101.json')
dnf300_2 = pd.read_json('../evaluation_set/deepnofakes/TRAIN/Set2_99.json')
dnf300_3 = pd.read_json('../evaluation_set/deepnofakes/TRAIN/Set3_100.json').drop(columns=['reason '])
```

```python
frames = [dnf300_1, dnf300_2, dnf300_3]
result = pd.concat(frames, sort=False)
```

```python
dnf300_1.keys(),dnf300_2.keys(),dnf300_3.keys()
```

```python
result.index = result.id
result.T.to_json('../evaluation_set/deepnofakes/dnf_300/combined_300.json')
```

```python
result.head(2)
```

```python
result.shape
```

# Pickle up sentence embedding and list

```python

id_sent_array = {}
id_sent_vector_array = {}
for idx in tqdm_notebook(result.id.values):
    with open('../evaluation_set/deepnofakes/TRAIN/cleaned/article_{0}.txt'.format(idx)) as f:
        article = f.readlines()
    article=''.join(article)
#     article = article.replace("'","").replace('"','').replace("‘",'').replace("’",'').replace("”",'').replace("“",'')
#     print(article)
    doc = sentencizer(article)
    sentences = [sent.string.strip() for sent in doc.sents]
#     print(sentences)
    id_sent_array[idx] = sentences
    sent_vectors = []
    for s in sentences:
        sent_vectors.append(nlp(s).vector)
    id_sent_vector_array[idx] = np.array(sent_vectors)
#     print(np.array(sent_vectors).shape)
#     break
```

```python
id_sent_array_file = '../evaluation_set/deepnofakes/dnf_300/cleaned/cleaned_dnf300_sent_array_id.p'
outfile = open(id_sent_array_file,'wb')
pickle.dump(id_sent_array,outfile)
outfile.close()
```

```python
id_sent_vector_array_file = '../evaluation_set/deepnofakes/dnf_300/cleaned/cleaned_dnf300_sent_vector_array_id.p'
outfile = open(id_sent_vector_array_file,'wb')
pickle.dump(id_sent_vector_array,outfile)
outfile.close()
```

```python
dnf700.keys()
```

```python
id_sent_array = {}
id_sent_vector_array = {}
for idx in tqdm_notebook(dnf700.id.values):
    with open('../evaluation_set/deepnofakes/TRAIN/700/article_{0}.txt'.format(idx)) as f:
        article = f.readlines()
    article=''.join(article)
    article = article.replace('[REF]','')
    doc = sentencizer(article)
    sentences = [sent.string.strip() for sent in doc.sents]
    id_sent_array[idx] = sentences
    sent_vectors = []
    for s in sentences:
        sent_vectors.append(nlp(s).vector)
    id_sent_vector_array[idx] = sent_vectors
```

```python
id_sent_array_file = '../evaluation_set/deepnofakes/dnf_700/dnf700_sent_array_id.p'
outfile = open(id_sent_array_file,'wb')
pickle.dump(id_sent_array,outfile)
outfile.close()
```

```python
id_sent_vector_array_file = '../evaluation_set/deepnofakes/dnf_700/dnf700_sent_vector_array_id.p'
outfile = open(id_sent_vector_array_file,'wb')
pickle.dump(id_sent_vector_array,outfile)
outfile.close()
```
